```python
datasets = [load(dataset) for dataset in DATASET_NAMES]
```

```python
def perform_dbscan(dataset, eps=3, min_samples=2):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
    dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(dataset)
    labels = dbscan.labels_
    visualize(dataset, ax=axes[0])
    visualize(dataset, labels, ax=axes[1])
```

```python
def min_max(dataset):
    normalized_df = (dataset-dataset.min())/(dataset.max()-dataset.min())
    normalized_df.name = df.name
    return normalized_df
```

```python
perform_dbscan(datasets[0], eps=4)
plt.show()
perform_dbscan(datasets[0], eps=5)
plt.show()
perform_dbscan(datasets[0], eps=6)
plt.show()
```
